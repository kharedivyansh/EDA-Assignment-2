```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
from collections import Counter
import plotly.express as px
```

```python
df = pd.read_csv("/content/amazon.csv")
```

```python
df.tail()
```

```python
df.drop_duplicates(inplace = True)
```

```python
#1 What is the average rating for each product category?
df["rating"].unique()
```

```python
# Replace any unexpected non-numeric values with NaN
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

# Now you can calculate the mean
Avg_rating = df["rating"].mean()
```

```python
Avg_rating
#hence average rating is above 4 for each product
```

```python
#2 What are the top rating_count products by category?
top_products_per_category = df.sort_values('rating_count', ascending=False).groupby('category').head(1)
top_products_per_category
```

```python
#3 What is the distribution of discounted prices vs. actual prices?
# Convert price columns to numeric after removing the currency symbol
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)

# Plotting the distributions
plt.figure(figsize=(16, 8))

# Distribution Plot
plt.subplot(1, 2, 1)
sns.histplot(df['actual_price'], kde=True, color='blue', label='Actual Price', bins=30)
sns.histplot(df['discounted_price'], kde=True, color='red', label='Discounted Price', bins=30)
plt.legend()
plt.title('Distribution Plot of Actual and Discounted Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Box Plot
plt.subplot(1, 2, 2)
sns.boxplot(data=df[['actual_price', 'discounted_price']])
plt.title('Box Plot of Actual and Discounted Prices')
plt.xlabel('Price Type')
plt.ylabel('Price')

plt.tight_layout()
plt.show()
```

```python
#4 How does the average discount percentage vary across categories?
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)

# Group by category and calculate the average discount percentage
category_discount = df.groupby('category')['discount_percentage'].mean().reset_index()

# Sort by average discount percentage
category_discount = category_discount.sort_values(by='discount_percentage', ascending=False)

# Plotting the average discount percentage across categories
plt.figure(figsize=(14, 8))
sns.barplot(x='discount_percentage', y='category', data=category_discount, palette='viridis')
plt.title('Average Discount Percentage Across Categories')
plt.xlabel('Average Discount Percentage')
plt.ylabel('Category')
plt.show()
```

```python
#5 What are the most popular product names?
# Fill NaN values in rating_count with 0
df['rating_count'] = df['rating_count'].fillna('0')

# Convert rating_count to numeric after removing commas
df['rating_count'] = df['rating_count'].str.replace(',', '').astype(int)

# Group by product name and calculate the total rating count
product_popularity = df.groupby('product_name')['rating_count'].sum().reset_index()

# Sort by the number of ratings in descending order
product_popularity = product_popularity.sort_values(by='rating_count', ascending=False)

# Display the top 10 most popular products
top_products = product_popularity.head(10)

# Plotting the top 10 most popular products
plt.figure(figsize=(14, 8))
sns.barplot(x='rating_count', y='product_name', data=top_products, palette='viridis')
plt.title('Top 10 Most Popular Product Names')
plt.xlabel('Number of Ratings')
plt.ylabel('Product Name')
plt.show()
```

```python
#6. What are the most popular product keywords?
# Concatenate all product names into a single string
all_product_names = ' '.join(df['product_name'].astype(str))

# Split the string into individual words
words = all_product_names.split()

# Count the frequency of each word
word_counts = Counter(words)

# Convert the word counts to a DataFrame
word_counts_df = pd.DataFrame(word_counts.items(), columns=['keyword', 'count'])

# Sort the DataFrame by count in descending order
word_counts_df = word_counts_df.sort_values(by='count', ascending=False)

# Display the top 20 most popular keywords
top_keywords = word_counts_df.head(20)

# Plotting the top 20 most popular keywords
plt.figure(figsize=(14, 8))
sns.barplot(x='count', y='keyword', data=top_keywords, palette='viridis')
plt.title('Top 20 Most Popular Product Keywords')
plt.xlabel('Count')
plt.ylabel('Keyword')
plt.show()
```

```python
#8. What is the correlation between discounted_price and rating?
selected_columns = ["discounted_price","rating"]
df[selected_columns].corr()
```

```python
#9. What are the Top 5 categories based on the highest ratings?
# Calculate the average rating for each category
category_ratings = df.groupby('category')['rating'].mean().reset_index()

# Sort categories based on average rating (descending)
top_categories = category_ratings.sort_values(by='rating', ascending=False).head(5)  # Top 5 categories

# Display the top categories based on highest ratings
print(top_categories[['category', 'rating']])
```

```python
#5 What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.
# Sort by popularity in ascending order
least_popular_tracks = df1.sort_values(by='Popularity').head(5)

# Select relevant columns
least_popular_tracks_info = least_popular_tracks[['Artist', 'Track Name', 'Popularity']]

# Display the result
print(least_popular_tracks_info)
```

```python
#6 Among the top 5 most popular artists, which artist has the highest popularity on average?
 # Group by 'Artist' and calculate the average popularity
average_popularity = df1.groupby('Artist')['Popularity'].mean().reset_index()

# Sort by average popularity in descending order
top_artists = average_popularity.sort_values(by='Popularity', ascending=False).head(5)

# Display the result
print(top_artists)
```

```python
#7 For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.
# Group by 'Artist' and calculate the average popularity
average_popularity = df1.groupby('Artist')['Popularity'].mean().reset_index()

# Sort by average popularity in descending order
top_artists = average_popularity.sort_values(by='Popularity', ascending=False).head(5)

# Create an empty list to store the most popular tracks for each top artist
most_popular_tracks_list = []

# Find the most popular track for each of the top 5 artists
for artist in top_artists['Artist']:
    artist_tracks = df1[df1['Artist'] == artist]
    most_popular_track = artist_tracks.sort_values(by='Popularity', ascending=False).head(1)
    most_popular_tracks_list.append(most_popular_track[['Artist', 'Track Name', 'Popularity']])

# Concatenate the list into a single DataFrame
most_popular_tracks = pd.concat(most_popular_tracks_list)

# Display the result
print(most_popular_tracks)
```
